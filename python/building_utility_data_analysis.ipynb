```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.DataFrame.from_csv('data_2/bps_energy_consumption_2011_2012/2012 Raw-Table 1.csv', index_col=None)
```

```python
col_names = list(df.columns)
col_names[0] = 'organization'
col_names[1] = 'sector'
col_names[2] = 'subsector'
col_names[3] = 'operation_name'
col_names[4] = 'operation_type'
col_names[8] = 'tot_floor_area'
col_names[9] = 'floor_area_unit'
col_names[-2] = 'en_intensity_ekWh_sqft'
df.columns = col_names
```

Cleaning

```python
for i in df.index:
    
    if df['sector'][i] == 'Post-secondary Educational Institute':
        df['sector'][i] = 'Post-Secondary Educational Institute'
    elif df['sector'][i] == 'School Boards':
        df['sector'][i] = 'School Board'
        
    if df['subsector'][i] == 'college':
        df['subsector'][i] = 'College'
    elif df['subsector'][i] == 'School Boards':
        df['subsector'][i] = 'School Board'
    elif df['subsector'][i] == 'University ':
        df['subsector'][i] = 'University'
```

```python
df.to_csv('data_2/cleaned_2012_not_normalized.csv')
```

```python
for i in df_norm.index:
    if df_norm['subsector'][i] == 'University ':
        df_norm['subsector'][i] = 'University'
```

```python
def plot_norm_data(df_norm, subsector):    
    temp_series = df_norm.query("subsector == @subsector")['eWh_per_HDD_sqft'] 
    temp_series.hist(bins = 100, range=[0, 100], facecolor='gray')
```
