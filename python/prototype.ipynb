Get the distributions for each operation type

```python
op_dist = {}
for op_type in set(df.operation_type):
    op_dist[op_type] = list(df.query("operation_type == @op_type")['en_intensity_ekWh_sqft'])
```

Add the percentile rank to the main df

```python
df['en_intensity_ekWh_sqft_pct_rank'] = pd.Series(index=df.index)
```

```python
for i in df.index:
    op_type = df.operation_type[i]
    pct_rank = stats.percentileofscore(op_dist[op_type], df.en_intensity_ekWh_sqft[i])
    df.en_intensity_ekWh_sqft_pct_rank[i] = pct_rank
```

Note: a percentile rank of 91 means the builing is in the top 10% of energy users for buildings of its type

Build the plots

```python
def get_best_worst(df, organization):
    org_df = df.query("organization == @organization")
    best_i = org_df.en_intensity_ekWh_sqft_pct_rank.idxmin()
    worst_i = org_df.en_intensity_ekWh_sqft_pct_rank.idxmax()
    return best_i, worst_i

def get_pct_rank_statement(df, i, best_worst):
    text = 'Your ' + best_worst + ' building is: ' + df.operation_name[i] + ' at '
    text += df.Address[i] + '\n which is in the '
    val = df.en_intensity_ekWh_sqft_pct_rank[i]
    if val <= 50.:
        text += 'top {0:0.0f}% '.format(val)
    else:
        text += 'bottom {0:0.0f}% '.format(100. - val)
    text += 'of ' + df.operation_type[i] + ' for energy conservation in Ontario.'
    return text

def get_org_statements(df, organization):
    best_i, worst_i = get_best_worst(df, organization)
    best_statement = get_pct_rank_statement(df, best_i, 'best')
    worst_statement = get_pct_rank_statement(df, worst_i, 'worst')
    n_properties = len(df.query("organization == @organization"))
    print organization + ' has ' + str(n_properties) + ' properties.\n'
    print best_statement
    print
    print worst_statement
    return best_i, worst_i
```

#### Now produce a histogram for a given building index

```python
def get_dist(df, i):
    op_type = df.operation_type[i]
    val = df.en_intensity_ekWh_sqft[i]
    f, ax = plt.subplots(1, 1, figsize=(5,5))
    ax.hist(op_dist[op_type], bins=100, color='grey', edgecolor='grey');
    ax.vlines(val, ax.get_ybound()[0], ax.get_ybound()[1], colors='red')
    ax.set_xlabel('Energy Intensity (ekWh/sqft)')
    ax.set_ylabel('Number of buildings')
    ax.set_title(op_type)
```
